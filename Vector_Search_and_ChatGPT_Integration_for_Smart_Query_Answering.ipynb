```python
from transformers import BertTokenizer, BertModel
import torch

# Load pre-trained BERT model and tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

# Example text input
text = "How do I reset my password?"

# Tokenize the input text
inputs = tokenizer(text, return_tensors='pt', padding=True, truncation=True)

# Get the embeddings (last hidden state of the model)
with torch.no_grad():
    outputs = model(**inputs)

# Get the embeddings from the last hidden state
embeddings = outputs.last_hidden_state.mean(dim=1).squeeze().numpy()  # Averaging over tokens

print("Generated Text Embedding:", embeddings)
```

```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Load pre-trained GPT-2 model and tokenizer
model = GPT2LMHeadModel.from_pretrained('gpt2')
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')

# Input text
input_text = "What is the capital of France?"

# Tokenize the input
input_ids = tokenizer.encode(input_text, return_tensors='pt')

# Generate a response from GPT-2
outputs = model.generate(input_ids, max_length=100, num_return_sequences=1, no_repeat_ngram_size=2)

# Decode the generated response
generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

print("Generated Text:", generated_text)
```

```python
import faiss
import numpy as np
from transformers import BertTokenizer, BertModel
import torch

# Set up BERT for embedding generation
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

# Set up FAISS for vector search
dimension = 768  # Correct dimension for 'bert-base-uncased' embeddings
index = faiss.IndexFlatL2(dimension)

# Add sample vectors (use embeddings from text input)
texts = ["How to reset password?", "What is AI?", "How to use Google Colab?"]

# Generate embeddings for each text
def generate_embedding(text):
    inputs = tokenizer(text, return_tensors='pt', padding=True, truncation=True)
    with torch.no_grad():
        outputs = model(**inputs)
    return outputs.last_hidden_state.mean(dim=1).squeeze().numpy()

embeddings = [generate_embedding(text) for text in texts]
index.add(np.array(embeddings, dtype='float32'))

# Query and search the index for nearest neighbors
query_text = "Tell me about AI"
query_embedding = generate_embedding(query_text)

D, I = index.search(np.array([query_embedding], dtype='float32'), k=3)

# Print results (nearest neighbors)
print("Nearest Neighbors Indices:", I)
print("Nearest Neighbors Texts:", [texts[i] for i in I[0]])

# Optional: Generate a response using GPT-2 based on the query
from transformers import GPT2LMHeadModel, GPT2Tokenizer

gpt2_model = GPT2LMHeadModel.from_pretrained('gpt2')
gpt2_tokenizer = GPT2Tokenizer.from_pretrained('gpt2')

input_ids = gpt2_tokenizer.encode(query_text, return_tensors='pt')
output = gpt2_model.generate(input_ids, max_length=100, num_return_sequences=1)

generated_text = gpt2_tokenizer.decode(output[0], skip_special_tokens=True)
print("GPT-2 Response:", generated_text)
```
